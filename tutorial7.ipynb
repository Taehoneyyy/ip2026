```python
# List Comprehensions 2

li = [3, 6, 2, 7]
result = [elem * 2 for elem in li]

print(result)
```

```python
# List Comprehensions 3

li = [('a', 1), ('b', 2), ('c', 7)]
result = [n * 3 for (x, n) in li]

print(result)
```

```python
# Filtered List Comprehension 2

li = [3, 6, 2, 7, 1, 9]
result = [elem * 2 for elem in li if elem > 4]

print(result)
```

```python
# Nested List Comprehensions

li = [3, 2, 4, 1]
result = [elem * 2 for elem in [item + 1 for item in li]]

print(result)
```

```python
# For Loops / List Comprehensions 비교

li = [3, 6, 2, 7]

# for문 방식
result_for = []
for elem in li:
    result_for.append(elem * 2)

print("for loop:", result_for)

# list comprehension 방식
result_comp = [elem * 2 for elem in li]

print("list comprehension:", result_comp)
```
