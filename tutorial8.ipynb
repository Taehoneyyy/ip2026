```python
# Calling a Function

def myfun(x, y):
    return x * y

print(myfun(3, 4))
```

```python
# Functions without returns

def show_message():
    print("Hello Python")

result = show_message()
print(result)
```

```python
# Functions are first-class objects in Python

def myfun(x):
    return x * 3

def apply(q, x):
    return q(x)

print(apply(myfun, 7))
```

```python
# Lambda Notation

def apply(q, x):
    return q(x)

print(apply(lambda z: z * 4, 7))
```

```python
# Default Values for Arguments

def myfun(b, c=3, d="hello"):
    return b + c

print(myfun(5, 3, "hello"))
print(myfun(5, 3))
print(myfun(5))
```

```python
# Keyword Arguments

def myfun(a, b, c):
    return a - b

print(myfun(2, 1, 43))
print(myfun(c=43, b=1, a=2))
print(myfun(2, c=43, b=1))
```
